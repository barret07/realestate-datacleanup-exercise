#### Exercise 16. Calculate the sample mean and variance of the variables: price, rooms, surface area and bathrooms (you must use the subset obtained in the question 14) (★★★)

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt

new_ds['normalized_price'] = new_ds.groupby('level5')['price'].transform(lambda x: (x - x.min()) / (x.max() - x.min()))

for population, data in new_ds.groupby('level5'):
    plt.hist(data['normalized_price'], bins=20, alpha=0.5, label=population)

plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.title('Normalized Prices Histogram')
plt.legend(title='Population')
plt.show()
```

**TODO: Markdown**. 
The lowest and most frequent prices are in Alcorcón, followed by those in Getafe. In Leganés, good prices can also be found but less frequently, while Fuenlabrada has a peak in frequency that diminishes as the price range increases.

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in the question 14) (★★★★)

Execute the following cell and then start coding in the next one. You must implement a simple code that transform the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
from ipyleaflet import Map, basemaps, Marker, CircleMarker
from IPython.display import display

# Map centred on (60 degrees latitude et -2.2 degrees longitude)
# Latitude, longitude
map = Map(center = (40.4168, -3.7038), zoom = 10, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
## HERE: plot the coordinates of the estates

## PUT HERE YOUR CODE:

print(south_belt[['latitude', 'longitude']].dtypes)

south_belt2 = south_belt.copy()
south_belt2['latitude'] = south_belt2['latitude'].str.replace(',','.').astype('float64')
south_belt2['longitude'] = south_belt2['longitude'].str.replace(',','.').astype('float64')

population_colours = {
    "Fuenlabrada": "red",
    "Leganés": "blue",
    "Getafe": "green",
    "Alcorcón": "purple"
}

# Iterate through the DataFrame and add markers to the map
for _, row in south_belt2[['latitude','longitude','level5']].iterrows():
  latitude = row['latitude']
  longitude = row['longitude']
  population = row['level5']
  color = population_colours.get(population, "gray")  # Default to gray if population not found

  marker = CircleMarker(location=(latitude, longitude), radius=5, color=color, fill=True, fill_color=color)
  map.add_layer(marker)

# Display the map
display(map)
```
